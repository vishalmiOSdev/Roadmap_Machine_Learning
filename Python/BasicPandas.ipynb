```python
df=pd.DataFrame(data=np.arange(0,9).reshape(3,3),index=["row1","row2","row3"],columns=["col1","col2","col3"])
print(df)
df[["col1","col2","col3"]]

```

```python
# using column  indexing
df=pd.DataFrame(data=np.arange(0,9).reshape(3,3),index=["row1","row2","row3"],columns=["col1","col2","col3"])
df[["col1","col2"]]


```

```python
# using row indexing
df=pd.DataFrame(data=np.arange(0,9).reshape(3,3),index=["row1","row2","row3"],columns=["col1","col2","col3"])
print(df)
df.loc[["row1","row2"]]


```

```python
# using row indexing
df=pd.DataFrame(data=np.arange(0,9).reshape(3,3),index=["row1","row2","row3"],columns=["col1","col2","col3"])
df.iloc[1:,1:]
```
